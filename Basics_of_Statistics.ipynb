#1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Disscuss nominal, ordinal, interval and ratio scales.
->
- Qualitative Data : Descriptive data that represents categories or labels.
  - Example : Colors of cars (red, blue, black) , types of cuisines (Indian , Chinese , Italian).
  
  - Ordinal Scale : Data categorized with a meaningful order but without equal intervals. Example : Movie ratings (poor , average , excellent ).

  - Nominal Scale : Data categorized without a specific order. Example: Gender (male, female ), types of fruits (apple, banana).

- Quantitative Data : Numerical data representing measurements or counts.
  - Example : Height of students (in cm), numbers of apples in a basket.

  - Interval Scale : Data with equal intervals but no true zero. Example : Temprature in Celsius or Fahrenheit.

  - Ratio Scale : Data with equal intervals and a true zero . Example : Weight (in kg) , time (in seconds).


#2. What are the measure of central tendency , and when should you use each? Discuss the mean , median, and mode with examples and situations where each is appropriate.
-> The measures of central tendency are statistical tools used to identify the center or typical value of a dataset. The Three main measures are:
- Mean : The average of a data set and the sum of all data values divided by the number of data points.
  - Use : When the data is continuous and does not have extreme outliers.
  - Example : Average marks of students in a class.

- Median : The middle value when the data is ordered.
  - Use : When data has outliers or skewed distribution.
  - Example : Median income in a city.

- Mode : The most frequently occuring value.
  - Use : When identifying the most common category or value.
  - Example : Most preferred brand in a survey.

#3. Explain the concept of dispersion. How do variannce and standard deviation measure the spread of data?
-> Dispersion : The measure of how data points spread around a central value.
  - Variance : The average of squared differences from the mean. It measures the overall spread of the data.

  - Standard Deviation : The square root of variance , indicating the average distance from the mean. It's in the same units as the data.
   - Example : Low standard deviation means data points are close to the mean ; high standard deviation indicates more spread.

#4. What is box plot, and what can it tell you about the distribution of data ?  
-> A box plot (or whisker plot) visually represents the distribution of data through five main points :    
  - Minimum
  - First quartile (Q1)
  - Median
  - Third quartile (Q3)
  - Maximum

- What it tells you:
  - Detects outliers.
  - Shows skewness and spread of the data.
  - Compares distribution between different data sets.


#5. Discuss the role of random sampling in making inferences about populations.
->
- Random Sampling :     
  - Ensures every individual in the population has an equal chance of being selected.
  - Reduces selection bias and makes the sample representative of the population.

- Role in inference :    
  - Helps generalize findings from the sample to the population.
  - Improves accuracy and reliability of statistical conclusions.
  - Example: Conducting a survey on 100 randomly selected citizens to predict election results.

#6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data ?
->
Skewess : A measure of the asymmetry in the distribution of data.
  - Types:

    - Positive Skew : Tail on the right, most data points are on the left. Example : Income distribution in  a city.
    - Negative Skew : Tail on the left , most data points are on the  right. Example : Age at retirement.
    - Symmetrical Distribution : No skewness; data is evenly distributed.

  - Effect on Interpretation :    
    - Skewness affects the choice of measures of central tendency.
    - Positively skewed data may require the use of the median instead of the mean for central tendency.
    - Skewness highlights the presence of outliers and deviations from normal distribution.

#7. What is the interquartile range (IQR), and how is it used to detect outliers?
-> IQR : Measures the spread of the middle 50 % of data.
   - Formula :    
      IQR = Q3 - Q1
- Outlier Detection :    
  - Outliers are values outside the range :    

Lower Bound = Q1 - 1.5 * IQR
Upper Bound = Q3 + 1.5 * IQR

#8. Disscuss the conditions under which the binomial distribution is used.
->
- Conditions for Binomial Distribution:
   a. The experiment has a fixed number of trials (n).
   b. Each trial has only two outcomes (success or failure).
   c. The probabilty of success (p) remains constant across trials.
   d. The trials are independent.
- Example : Tossing a coin 10 times and counting the number of heads.

#9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
->
- Properties of Normal Distribution :    
  - Symmetrical and bell shaped.
  - Mean, Median, and mode are equal.
  - Defined by mean (µ) and standard deviation (σ)

- Empirical Rule :     
  - 68% of data lies within 1 standard deviation ($µ\pmσ$).
  - 95% of data lies within 2 standard deviations (  $µ\pm2σ$).
  - 99.7% of data lies within 3 standard deviation ($µ\pm3σ$).

#10. Provide a real - life example of a Poisson process and calculate the probability or a specific event.

-> Poisson Process : Models the number of events occuring in a fixed interval of time or space when events happen independently.
- Example : Number of customer arrivals at a bank in an hour.
 - if the average is 5 customers /hour( λ = 5 ), the probability of exactly 3 arrivals in an hour is calculated as:

 $$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$$

 Where:
 - k: Number of events
 - λ: Average number of events per unit interval
 - e: Eular's number( ≈ 2.718)

 $$P(X=3) = \frac{5^3 e^{-5}}{3!}$$
